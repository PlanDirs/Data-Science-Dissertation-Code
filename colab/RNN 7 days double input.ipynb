#### construct the RNN model

```python
import torch
import torch.nn as nn
from scipy.stats import pearsonr

import torch
import torch.nn as nn

class MultiInputSingleOutputRNN(nn.Module):
    def __init__(self, input_size1, input_size2, hidden_size1, hidden_size2, output_size, time_step, num_layers1, num_layers2, dropout):
        super(MultiInputSingleOutputRNN, self).__init__()

        self.time_step = time_step

        self.hidden_size1 = hidden_size1
        self.hidden_size2 = hidden_size2

        self.num_layers1 = num_layers1
        self.num_layers2 = num_layers2

        self.rnn1 = nn.RNN(input_size1, hidden_size1, num_layers1, batch_first=True, dropout = dropout)
        self.rnn2 = nn.RNN(input_size2, hidden_size2, num_layers2, batch_first=True, dropout = dropout)

        self.fc1 = nn.Linear( (hidden_size1+hidden_size2) * self.time_step, output_size)
        #self.fc2 = nn.Linear((hidden_size1+hidden_size2), output_size)

    def forward(self, input1, input2):
        batch_size = input1.size(0)
        sequence_length = input1.size(1)

        # Process input1 with rnn1
        hidden1 = self.init_hidden(self.num_layers1, batch_size, self.hidden_size1)
        out1, hidden1 = self.rnn1(input1, hidden1)
        #print(out1.shape)

        # Process input2 with rnn2
        hidden2 = self.init_hidden(self.num_layers2, batch_size, self.hidden_size2)
        out2, hidden2 = self.rnn2(input2, hidden2)
        #print(out2.shape)

        # Concatenate the outputs of rnn1, rnn2, and rnn3 along the feature dimension
        combined_output = torch.cat((out1, out2), dim=2)
        #print(combined_output.shape)

        out = combined_output[:, -self.time_step:, :]
        #print(out.shape)

        #out = out.view(batch_size, -1)
        out = out.view(out.size(0), -1)
        #print(out.shape)

        out = self.fc1(out)

        #out = self.fc2(out)

        return out

    def init_hidden(self, num_layers, batch_size, hidden_size):
        return torch.zeros(num_layers, batch_size, hidden_size).to(device)
```

#### early stop function

```python
# 定义早停函数
def early_stopping(val_loss, patience=10, delta=0):
  if len(val_loss) < patience + 1:
    return False

  best_loss = min(val_loss[:-patience])
  last_loss = val_loss[-1]

  if last_loss > best_loss - delta:
    return True

  return False
```

#### define function for training and validation

```python
from torch.optim.lr_scheduler import StepLR

# 定义训练和验证过程
def train_and_validate_model(model, train_loader, val_loader, optimizer, criterion, num_epochs, patience=30, delta=0.005):
    train_loss_history = []
    val_loss_history = []
    best_model_state = None

    # 学习率衰减
    step_size = 15
    gamma = 0.95
    scheduler = StepLR(optimizer, step_size=step_size, gamma=gamma)

    # 训练模型
    for epoch in range(num_epochs):
        model.train()
        train_total_loss = 0.0
        for i, (X1_batch, X2_batch, y_batch) in enumerate(train_loader):
            # Clear gradients from the previous iteration
            optimizer.zero_grad()

            # Forward pass: compute model predictions
            predictions = model(X1_batch.to(device), X2_batch.to(device))
            #print('here is prediction：', predictions.shape)

            # Compute the loss
            loss = criterion(predictions, y_batch.to(device))
            #print('here is loss：', loss.item())

            # Backpropagation: compute gradients of the loss with respect to model parameters
            loss.backward()

            # Optimization: update model parameters using the gradients
            optimizer.step()

            # Accumulate the total loss for this epoch

            train_total_loss += loss.item()

            #print(total_loss)
        # Calculate the average loss for this epoch
        scheduler.step()
        train_loss = train_total_loss / len(train_loader)
        train_loss_history.append(train_loss)
        #print('toatl loss: ', total_loss)
        #print(len(train_loader))
        #print('average loss: ', average_loss)

        #if (epoch + 1) % 10 == 0:
        #    print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {average_loss:.4f}')

        model.eval()
        with torch.no_grad():
            val_total_loss = 0.0
            for batch_X1, batch_X2, batch_y in val_loader:
                predictions = model(batch_X1.to(device), batch_X2.to(device))
                loss = criterion(predictions, batch_y.to(device))
                val_total_loss += loss.item()
            validation_loss = val_total_loss / len(val_loader)
            val_loss_history.append(validation_loss)

        # 检查是否早停
        if early_stopping(val_loss_history, patience, delta):
            print("\n Early stopping triggered!")
            best_model_state = model.state_dict()
            break

    return best_model_state
```

#### reshape data

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

def reshape_data(X1, X2, y, num_channels):
    num_samples = X1.shape[0] - num_channels + 1
    channels_X1 = [X1.iloc[i:i+num_channels, :] for i in range(num_samples)]
    channels_X2 = [X2.iloc[i:i+num_channels, :] for i in range(num_samples)]

    X_reshaped = np.array(channels_X1).reshape(num_samples, num_channels, X1.shape[1])
    X2_reshaped = np.array(channels_X2).reshape(num_samples, num_channels, X2.shape[1])

    y_reshaped = y[-num_samples:]

    return X_reshaped, X2_reshaped, y_reshaped
```

```python
def RNN_fitness_function(params):

    # Defince the parameters
    hidden_size1 = int(params[0])
    hidden_size2 = int(params[1])
    num_layers1 = int(params[2])
    num_layers2 = int(params[3])
    learning_rate = params[4]
    batch_size = [32, 64, 128, 256][int(params[5])]
    weight_decay = params[6]
    sequence_length = int(params[7])
    dropout = params[8]
    time_step = int(params[9])
    epoch = [500, 1000][int(params[10])]


    reshaped_X1, reshaped_X2, reshaped_y = reshape_data(X1, X2, y, sequence_length)

    # 划分训练集为新的训练集和验证集
    X1_train, X1_temp, X2_train, X2_temp, y_train, y_temp = train_test_split(reshaped_X1, reshaped_X2, reshaped_y, test_size=0.3, random_state=42)

    # 划分训练集为新的训练集和验证集
    X1_val, X1_test, X2_val, X2_test, y_val, y_test = train_test_split(X1_temp, X2_temp, y_temp, test_size=0.5, random_state=42)



    # 转换为 PyTorch 张量
    torch.manual_seed(42)  # set the random seed

    X1_train = torch.tensor(X1_train, dtype=torch.float32).to(device)
    X1_val = torch.tensor(X1_val, dtype=torch.float32).to(device)
    X1_test = torch.tensor(X1_test, dtype=torch.float32).to(device)

    X2_train = torch.tensor(X2_train, dtype=torch.float32).to(device)
    X2_val = torch.tensor(X2_val, dtype=torch.float32).to(device)
    X2_test = torch.tensor(X2_test, dtype=torch.float32).to(device)

    y_train = torch.tensor(y_train.values, dtype=torch.float32).to(device)
    y_val = torch.tensor(y_val.values, dtype=torch.float32).to(device)
    y_test = torch.tensor(y_test.values, dtype=torch.float32).to(device)


    train_dataset = torch.utils.data.TensorDataset(X1_train, X2_train, y_train)
    val_dataset = torch.utils.data.TensorDataset(X1_val, X2_val, y_val)
    test_dataset = torch.utils.data.TensorDataset(X1_test, X2_test, y_test)


    # Specify the batch size
    batch_size = batch_size

    # Create DataLoaders for training, validation, and testing
    train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = torch.utils.data.DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
    test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size, shuffle=False)



    # 初始化模型和优化器
    input_size1 = X1_train.shape[2]
    input_size2 = X2_train.shape[2]
    output_size = y_train.shape[1]
    #device = torch.device("cuda")
    model = MultiInputSingleOutputRNN(input_size1, input_size2, hidden_size1, hidden_size2, output_size, time_step, num_layers1, num_layers2, dropout).to(device)
    optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=weight_decay)
    criterion = nn.MSELoss()
    num_epochs = epoch

    # train!!!
    best_model_state = train_and_validate_model(model, train_loader, val_loader, optimizer, criterion, num_epochs=num_epochs, patience=30, delta=0.001)
    #print('this is :', best_model_state)
    #model.load_state_dict(best_model_state)
    model.eval()

    with torch.no_grad():
        total_mse = 0.0
        for batch_X1, batch_X2, batch_y in val_loader:
            # Forward pass: compute model predictions
            predictions = model(batch_X1.to(device), batch_X2.to(device))

            # Compute the loss (assuming you have defined the loss function as 'criterion')
            mse = criterion(predictions, batch_y.to(device))

            # Accumulate the total loss for this validation set
            total_mse += mse.item()

        # Calculate the average loss for the entire validation set
        #print(total_mse)
        mse_loss = total_mse / len(val_loader)

        # Print the average loss for this validation set
        #print(f"Validation Loss: {validation_loss:.4f}")

    return mse_loss
```

```python
def median_absolute_error(y_true, y_pred):
    absolute_errors = torch.abs(y_true - y_pred)
    return torch.median(absolute_errors).item()
```

```python
def pearson_correlation(y_true, y_pred):
    y_true_mean = torch.mean(y_true)
    y_pred_mean = torch.mean(y_pred)
    numerator = torch.sum((y_true - y_true_mean) * (y_pred - y_pred_mean))
    denominator = torch.sqrt(torch.sum((y_true - y_true_mean) ** 2)) * torch.sqrt(torch.sum((y_pred - y_pred_mean) ** 2))
    pearson_corr = numerator / denominator
    return pearson_corr.item()
```

```python
def index_of_agreement(y_true, y_pred):
    numerator = torch.sum((y_true - y_pred) ** 2)
    denominator = torch.sum((torch.abs(y_pred - torch.mean(y_true)) + torch.abs(y_true - torch.mean(y_true))) ** 2)
    index_agreement = 1 - (numerator / denominator)
    return index_agreement.item()
```
