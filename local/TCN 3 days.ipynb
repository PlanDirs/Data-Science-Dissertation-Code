```python
X = df.drop(['next_1_day_mean_temp','next_2_day_mean_temp','next_3_day_mean_temp'], axis=1)
y = df[['next_1_day_mean_temp', 'next_2_day_mean_temp', 'next_3_day_mean_temp']]
```

```python
# 假设您的 DataFrame 名为 df，形状为 (100, 22)

# 将数据划分为通道
num_channels = 10  # 每个通道包含的行数
num_samples = X.shape[0] - num_channels + 1
channels = []
for i in range(num_samples):
    channel = X.iloc[i:i+num_channels, :]
    channels.append(channel)

# 重新组织形状为 (x, x, x)
x = 10  # 每个通道中的行数
reshaped_X = np.array(channels).reshape(num_samples, x, X.shape[1])

# reshaped_df 的形状为 (num_samples, x, num_channels, num_features)
```

```python
reshaped_y = y[-num_samples:]
```

```python
# 转化张量

import torch
import torch.nn as nn

input_data = torch.tensor(reshaped_X, dtype=torch.float32)

reshaped_y = reshaped_y.values
target_data = torch.tensor(reshaped_y, dtype=torch.float32)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

data = input_data  # shape: [batch_size, channels, sequence_length].
targets = target_data
```

```python

# 定义TCN的残差块
class ResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size, stride, padding):
        super(ResidualBlock, self).__init__()
        self.conv1 = nn.Conv1d(in_channels, out_channels, kernel_size, stride=stride, padding=padding)
        self.activation1 = nn.ReLU()
        self.conv2 = nn.Conv1d(out_channels, out_channels, kernel_size, stride=stride, padding=padding)
        self.activation2 = nn.ReLU()

    def forward(self, x):
        residual = x
        out = self.conv1(x)
        out = self.activation1(out)
        out = self.conv2(out)
        out = self.activation2(out)
        out = torch.add(out, residual)  # 跳跃连接，将残差与原始输入相加
        return out

# 定义TCN
class TCN(nn.Module):
    def __init__(self, in_channels, num_residual_blocks, residual_channels, kernel_size, stride, padding):
        super(TCN, self).__init__()
        self.conv1 = nn.Conv1d(in_channels, residual_channels, kernel_size, stride=stride, padding=padding)
        self.activation1 = nn.ReLU()
        self.residual_blocks = nn.Sequential(
            *[ResidualBlock(residual_channels, residual_channels, kernel_size, stride, padding) 
              for _ in range(num_residual_blocks)]
        )
        self.fc = nn.Linear(residual_channels, 3)  # 回归问题的输出是一个单一值，因此使用1个输出单元

    def forward(self, x):
        #print('begin:', x.shape)
        x = x.permute(0, 2, 1)  # 调整通道维度的顺序
        #print('adjust:', x.shape)
        out = self.conv1(x)
        #print('conv1: ', out.shape)
        out = self.activation1(out)
        #print('activation: ', out.shape)
        out = self.residual_blocks(out)
        #print('residual: ', out.shape)
        out = torch.mean(out, dim=2)  # 对时间维度进行平均池化
        #print('pooling: ', out.shape)
        out = self.fc(out)
        #print('full connect:', out.shape)
        return out
```

```python
# 定义参数
in_channels = 22  # 输入通道数
num_residual_blocks = 2  # 残差块的数量
residual_channels = 200  # 残差块中卷积层的输出通道数
kernel_size = 3  # 卷积核大小
stride = 1  # 卷积步长
padding = 1  # 卷积填充大小

# 初始化TCN
tcn = TCN(in_channels, num_residual_blocks, residual_channels, kernel_size, stride, padding)

# 定义损失函数和优化器
criterion = nn.MSELoss()
optimizer = optim.Adam(tcn.parameters(), lr=0.0005)
```

```python
num_epochs = 100
for epoch in range(num_epochs):
    optimizer.zero_grad()
    outputs = tcn(data)
    loss = criterion(outputs, targets)
    loss.backward()
    optimizer.step()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")

print("Finished Training")

# 在测试集上进行预测
```
