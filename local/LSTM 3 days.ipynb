```python
# 假设您的 DataFrame 名为 df，形状为 (100, 22)

# 将数据划分为通道
num_channels = 10  # 每个通道包含的行数
num_samples = X.shape[0] - num_channels + 1
channels = []
for i in range(num_samples):
    channel = X.iloc[i:i+num_channels, :]
    channels.append(channel)

# 重新组织形状为 (x, x, x)
x = 10  # 每个通道中的行数
reshaped_X = np.array(channels).reshape(num_samples, x, X.shape[1])

# reshaped_df 的形状为 (num_samples, x, num_channels, num_features)
```

```python
reshaped_y = y[-num_samples:]
```

```python
# 转化张量

import torch
import torch.nn as nn

input_data = torch.tensor(reshaped_X, dtype=torch.float32)

reshaped_y = reshaped_y.values
target_data = torch.tensor(reshaped_y, dtype=torch.float32)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from scipy.stats import pearsonr

# 定义LSTM模型
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size, time_step, dropout):
        super(LSTMModel, self).__init__()
        
        self.hidden_size = hidden_size
        
        self.time_step = time_step
        
        self.num_layers = num_layers
        
        # LSTM层
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True, dropout=dropout)

        # 整合
        self.fc1 = nn.Linear(hidden_size * self.time_step, hidden_size)
        
        # 输出层
        self.fc2 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        
        
        out, _ = self.lstm(x, (h0, c0))
        out = out[:, -self.time_step:, :]
        
        out = out.reshape(out.size(0), -1)
        
        out = self.fc1(out)
        out = self.fc2(out)
        return out
```

```python


# 假设输入维度为10，隐藏状态维度为20，LSTM层数为2，输出维度为1
input_size = 22
hidden_size = 50
num_layers = 3
output_size = 3
time_step = 5
dropout = 0.1


# 创建LSTM模型
lstm_model = LSTMModel(input_size, hidden_size, num_layers, output_size, time_step, dropout)


# 定义损失函数和优化器
criterion = nn.MSELoss()
optimizer = optim.Adam(lstm_model.parameters(), lr=0.003)

# 生成随机输入数据和目标值（假设输入序列长度为50，批次大小为32）
#sequence_length = 10
#batch_size = 15313
#input_data = torch.randn(batch_size, sequence_length, input_size)
#target_data = torch.randn(batch_size, output_size)

# 训练模型
num_epochs = 100
for epoch in range(num_epochs):
    lstm_model.train()
    optimizer.zero_grad()

    # 前向传播
    output = lstm_model(input_data)

    # 计算损失
    loss = criterion(output, target_data)

    # 反向传播和优化
    loss.backward()
    optimizer.step()

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item()}")

    
# 预测
with torch.no_grad():
    #test_input = torch.randn(1, sequence_length, input_size)
    predictions = lstm_model(input_data)
    # 计算MAE和MSE
    test_mse = nn.MSELoss()(predictions, target_data)
    
    print(f'MSE: {test_mse.item():.4f}')
```

```python
test_mae = nn.L1Loss()(predictions, target_data)
    
    test_rmse = torch.sqrt(test_mse)
    
    absolute_errors = torch.abs(predictions - target_data)
    test_mdae = torch.median(absolute_errors)
    
    percentage_errors = absolute_errors / target_data
    percentage_errors = percentage_errors[~torch.isinf(percentage_errors) & ~torch.isnan(percentage_errors)]
    test_mape = torch.mean(torch.abs(percentage_errors)) * 1000
    
    squared_percentage_errors = torch.pow(percentage_errors * 100, 2)
    squared_percentage_errors = squared_percentage_errors[~torch.isinf(squared_percentage_errors)]
    mean_squared_percentage_error = torch.mean(squared_percentage_errors)
    test_rmspe = torch.sqrt(mean_squared_percentage_error)
    
    batch_size = 100
    num_batches = predictions.shape[0] // batch_size
    pearson_coeffs = []
    for i in range(num_batches):
        start_idx = i * batch_size
        end_idx = (i + 1) * batch_size

        predictions_batch = predictions[start_idx:end_idx].numpy().flatten()
        target_data_batch = target_data[start_idx:end_idx].numpy().flatten()
        pearson_coefficients, _ = pearsonr(predictions_batch, target_data_batch)
        pearson_coeffs.append(pearson_coefficients)
    pearson_coefficient = np.mean(pearson_coeffs)
    
    def index_of_agreement(predictions, observations):
        numerator = torch.sum(torch.square(observations - predictions))
        denominator = torch.sum(torch.square(torch.abs(predictions - torch.mean(observations)) + torch.abs(observations - torch.mean(observations))))
        ioa = 1 - (numerator / denominator)
        return ioa
    
    ioa = index_of_agreement(predictions, target_data)
    
    print(f'MSE: {test_mse.item():.4f}')
    print(f'RMSE: {test_rmse.item():.4f}')
    print(f'MAE: {test_mae.item():.4f}')
    print(f'MdAE: {test_mdae.item():.4f}')
    print(f'MAPE: {test_mape.item():.4f}')
    print(f'RMSPE: {test_rmspe.item():.4f}')
    print(f'Pearson Coefficient: {pearson_coefficient}')
    print(f'Index of Agreement (IoA): {ioa:.4f}')
```
