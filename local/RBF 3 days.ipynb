```python
import torch.nn.init as init

torch.manual_seed(42)

class RBFN(nn.Module):
    """
    以高斯核作为径向基函数
    """
    def __init__(self, centers, beta, n_out=3):
        """
        :param centers: shape=[center_num,data_dim]
        :param n_out:
        """
        super(RBFN, self).__init__()
        self.n_out = n_out   
        self.num_centers = centers.size(0)  # 隐层节点的个数  RBF神经元个数
        
        self.centers = nn.Parameter(centers) # 设置RBF神经元的中心点
        self.beta = nn.Parameter(beta)
        #self.beta = torch.ones(1, self.num_centers)   # 设置RBF神经元的方差
        self.linear = nn.Linear(self.num_centers, self.n_out, bias=True)   
        # 初始化权重
        #init.xavier_normal_(self.linear.weight)
        init.kaiming_normal_(self.linear.weight, nonlinearity='relu')

 
 
    def kernel_fun(self, batches):
        n_input = batches.size(0)  # number of inputs，0代表行数
        A = self.centers.view(self.num_centers, -1).repeat(n_input, 1, 1)  #view成num_centers行，然后再重复n_input次
        B = batches.view(n_input, -1).unsqueeze(1).repeat(1, self.num_centers, 1)  
        C = torch.exp(-self.beta.mul((A - B).pow(2).sum(2, keepdim=False)))

        return C
 
    def forward(self, batches):
        out = self.kernel_fun(batches)
        out = self.linear(out)
        
        return out
```

```python
import torch
import numpy as np
from sklearn.cluster import KMeans

def kmeans_rbf_params(X_train, number_of_centers):
    """
    使用KMeans算法对数据进行聚类，并返回RBFNN的中心点(cluster_centers)和beta值(betas)。

    参数：
    X_train (torch.Tensor): 训练数据，形状为[样本数, 特征维度]。
    number_of_centers (int): RBF神经元的数量。

    返回：
    cluster_centers (np.ndarray): RBF神经元的中心点，形状为[数量, 特征维度]。
    betas (list): 每个RBF神经元的beta值列表。
    """
    # 设置随机种子
    torch.manual_seed(42)

    # 随机生成RBFNN的中心点
    random_centers = torch.randn(number_of_centers, X_train.shape[1], dtype=torch.float32)

    # 使用KMeans聚类
    #kmeans = KMeans(n_clusters=number_of_centers, init=random_centers.numpy(), n_init=1)
    kmeans = KMeans(n_clusters=number_of_centers, n_init=20)
    kmeans.fit(X_train)

    # 获取聚类中心
    cluster_centers = kmeans.cluster_centers_

    # 将X_train转换为numpy数组
    X_train_np = X_train.to_numpy()

    # 初始化beta
    betas = []
    for i in range(number_of_centers):
        cluster_indices = np.where(kmeans.labels_ == i)[0]
        distances = np.linalg.norm(X_train_np[cluster_indices] - cluster_centers[i], axis=1)
        beta_i = np.mean(distances)
        betas.append(beta_i)

    return cluster_centers, betas
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np

class RBFN(nn.Module):
    def __init__(self, num_centers, num_outputs):
        super(RBFN, self).__init__()
        self.num_centers = num_centers
        self.num_outputs = num_outputs
        
        # 定义RBF神经元的中心点，作为模型的可学习参数
        self.centers = nn.Parameter(torch.randn(num_centers, 1))  # 这里我们假设输入数据为一维
        
        # 定义RBF神经元的方差，作为模型的可学习参数
        self.beta = nn.Parameter(torch.ones(1, num_centers))  # 这里我们假设所有RBF神经元的方差相同
        
        # 定义输出层
        self.linear = nn.Linear(num_centers, num_outputs, bias=True)
        
    def kernel_fun(self, batches):
        A = self.centers.view(self.num_centers, -1)
        B = batches.view(-1, 1)
        C = torch.exp(-self.beta.mul((A.unsqueeze(1) - B).pow(2)))
        return C
    
    def forward(self, batches):
        radial_val = self.kernel_fun(batches)
        output = self.linear(radial_val)
        return output
```

```python
# 定义RBF神经元的数量和输出层的节点数
num_centers = 10
num_outputs = 1

# 创建RBFN实例
rbfn = RBFN(num_centers, num_outputs)

# 定义输入数据和目标数据
input_data = torch.linspace(0, 1, 100).view(-1, 1)  # 100个一维数据点
target_data = torch.sin(input_data * 5 * np.pi)  # 目标函数为正弦函数

# 定义损失函数和优化器
criterion = nn.MSELoss()
optimizer = optim.Adam(rbfn.parameters(), lr=0.1)

# 训练RBFN模型
num_epochs = 1000
for epoch in range(num_epochs):
    rbfn.zero_grad()
    outputs = rbfn(input_data)
    loss = criterion(outputs, target_data)
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 100 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}')

# 测试模型
test_input = torch.linspace(0, 1, 20).view(-1, 1)
with torch.no_grad():
    predictions = rbfn(test_input)

# 打印预测结果
print("Input  | Target | Prediction")
for i in range(test_input.size(0)):
    print(f'{test_input[i][0]:.2f} | {target_data[i][0]:.4f} | {predictions[i][0]:.4f}')
```
