```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
data = {'Method':['RF', 'SVR', 'XGBoost', 'MLP', 'RBFNN', 'RNN', 'D-RNN', 'LSTM', 'Bi-LSTM', 'TCN', 'Transformer'],
        '3 Days': [3.8140, 3.8891, 3.8853, 3.8713, 3.9758, 3.6094, 3.5664, 3.5532, 3.5970, 2.3759, 2.9144],
        '7 Days': [6.3880, 6.6225, 6.4001, 6.3918, 6.5255, 6.4271, 6.1578, 1.4682, 6.1485, 2.1954, 2.3268],
        '15 Days': [8.002, 8.2072, 8.0892, 8.4423, 8.4695, 4.4770, 7.4815, 2.1345, 1.6624, 1.1006, 2.2315],
        '30 Days': [8.5587, 9.286, 8.7411, 8.5891, 9.5360, 4.6222, 8.0288, 3.1599, 2.9562, 4.8344, 1.9495]}
```

```python
df = pd.DataFrame(data)
```

```python
import seaborn as sns

#  deep 
colors = sns.color_palette("deep", 11)

plt.figure(figsize=(10, 6))

for i, method in enumerate(df['Method']):
    plt.plot(df.columns[1:], df.iloc[i, 1:], marker='o', label=method, color=colors[i], linewidth=2.5, linestyle='-')  #  linewidth  linestyle 

plt.title('A Comprehensive Study of Multi-Method Prediction Performance', fontsize=16)  # 
plt.xlabel('Predicting Range', fontsize=15)  # X
plt.ylabel('MSE', fontsize=12)  # Y
plt.grid(True, linestyle='-', linewidth=0.5, alpha=0.7)
plt.legend(loc='best')  # Matplotlib
plt.tight_layout()

# 
#ax = plt.gca()
#ax.set_facecolor('#F0F0F0')  # 

plt.show()
```
