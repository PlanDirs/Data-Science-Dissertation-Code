```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### 15 days

```python
df_BiLSTM15day_EC = pd.read_csv("C:/Users/Administrator/Desktop/JN/DS Project/Bi-LSTM15day_EC.csv")
```

```python
selected_columns = ['group', 'Date', 'Next_15_Days_Energy_Total', 'Predictions']
df_BiLSTM15day_EC = df_BiLSTM15day_EC[selected_columns]
```

```python
df_BiLSTM15day_EC.rename(columns={'Predictions': 'BiLSTM_Predictions'}, inplace=True)
```

```python
df_BiLSTM15day_EC['BiLSTM_Predictions'] = df_BiLSTM15day_EC['BiLSTM_Predictions'].str.strip('[]').astype(float)
```

```python
df_BiLSTM15day_EC
```

```python
df_TCN15day_EC = pd.read_csv("C:/Users/Administrator/Desktop/JN/DS Project/TCN15day_EC.csv")
```

```python
selected_columns = ['group', 'Date', 'Next_15_Days_Energy_Total', 'Predictions']
df_TCN15day_EC = df_TCN15day_EC[selected_columns]
```

```python
df_TCN15day_EC.rename(columns={'Predictions': 'TCN_Predictions'}, inplace=True)
```

```python
df_TCN15day_EC['TCN_Predictions'] = df_TCN15day_EC['TCN_Predictions'].str.strip('[]').astype(float)
```

```python
df_TCN15day_EC
```

```python
df_TCN15day_EC['TCN_Predictions'][0]
```

```python
import matplotlib.pyplot as plt

predicted_values_BiLSTM = df_BiLSTM15day_EC['BiLSTM_Predictions']  
predicted_values_TCN = df_TCN15day_EC['TCN_Predictions']

true_values_BiLSTM = df_BiLSTM15day_EC['Next_15_Days_Energy_Total']
true_values_TCN = df_TCN15day_EC['Next_15_Days_Energy_Total']

errors_BiLSTM = [predicted - true for predicted, true in zip(predicted_values_BiLSTM, true_values_BiLSTM)]
errors_TCN = [predicted - true for predicted, true in zip(predicted_values_TCN, true_values_TCN)]

plt.figure(figsize=(10, 6))

color_palette = ['#9467bd', '#ffeb3b']
# ['#1f77b4', '#d62728']

# 使用透明度来减少颜色之间的遮挡
plt.scatter(true_values_BiLSTM, errors_BiLSTM, color=color_palette[0], label='BiLSTM', alpha=0.25, s=20)
plt.scatter(true_values_TCN, errors_TCN, color=color_palette[1], label='TCN', alpha=0.3, s=20)
plt.axhline(y=0, color='black', linestyle='--')

plt.xlabel('True Values')
plt.ylabel('Errors (Predicted - True)')
plt.title('Residuals Comparison between BiLSTM and TCN Models')
plt.legend()

plt.grid(True, linestyle='--', alpha=0.7)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
#ax.set_facecolor('#f5f5f5')

plt.tight_layout()
plt.show()
```
