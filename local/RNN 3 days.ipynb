```python
import torch
import torch.nn as nn
from scipy.stats import pearsonr

class SingleInputSingleOutputRNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, time_step, num_layers, dropout):
        super(SingleInputSingleOutputRNN, self).__init__()

        self.time_step = time_step

        self.hidden_size = hidden_size

        self.num_layers = num_layers

        self.rnn = nn.RNN(input_size, hidden_size, num_layers, batch_first=True, dropout=dropout)
        #self.rnn = nn.RNN(input_size, hidden_size, num_layers, dropout, batch_first=True)
        #self.fc1 = nn.Linear(hidden_size, 32)
        #self.fc2 = nn.Linear(32, output_size)
        self.fc1 = nn.Linear(time_step * hidden_size, hidden_size)
        self.fc2 = nn.Linear(hidden_size, output_size)
        
    def forward(self, x):
        batch_size = x.size(0)
        sequence_length = x.size(1)

        hidden = self.init_hidden(batch_size) # initialize the hidden state 

        out, hidden = self.rnn(x, hidden)
        #time_step = 3
        out = out[:, -self.time_step:, :]  # 选择最后一个时间步的输出作为最终的输出结果
        #print(out.shape)
        
        out = out.reshape(out.size(0), -1)
        #print(out.shape)
        
        out = self.fc1(out)
        out = self.fc2(out)
        #out = torch.relu(out)
        #out = self.fc2(out)

        return out

    def init_hidden(self, batch_size):
        return torch.zeros(self.num_layers, batch_size, self.hidden_size)

# Define the input tensor
#input_data = torch.randn(15, 10, 22)  # Batch size: 1500, Sequence length: 10, Input size: 22
#target = torch.randn(15, 3)  # Batch size: 15, 3个目标值

```

```python
# 定义早停函数
def early_stopping(val_loss, patience=10, delta=0):
    if len(val_loss) < patience + 1:
        return False

    best_loss = min(val_loss[:-patience])
    last_loss = val_loss[-1]

    if last_loss > best_loss - delta:
        return True

    return False
```

```python
from torch.optim.lr_scheduler import StepLR

# 定义模型
model = SingleInputSingleOutputRNN(input_size=22, hidden_size=25, output_size=3, time_step=1, num_layers = 1, dropout = 1.03179779e-02)

# 定义损失函数
criterion = nn.MSELoss()  # 使用均方误差损失函数

# 定义优化器
optimizer = torch.optim.Adam(model.parameters(), lr=8.68102041e-04, weight_decay = 5.77752695e-03)

# 训练模型
num_epochs = 500

train_loss_history = []
val_loss_history = []
best_model_state = None

# 学习率衰减
step_size = 15
gamma = 0.9
scheduler = StepLR(optimizer, step_size=step_size, gamma=gamma)

# 训练模型
for epoch in range(num_epochs):
    model.train()
    train_total_loss = 0.0
    for i, (X_batch, y_batch) in enumerate(train_loader):
        # Clear gradients from the previous iteration
        optimizer.zero_grad()

        # Forward pass: compute model predictions
        predictions = model(X_batch)
        #print('here is prediction：', predictions.shape)

        # Compute the loss
        loss = criterion(predictions, y_batch)
        #print('here is loss：', loss.item())

        # Backpropagation: compute gradients of the loss with respect to model parameters
        loss.backward()

        # Optimization: update model parameters using the gradients
        optimizer.step()

        # Accumulate the total loss for this epoch

        train_total_loss += loss.item()

        #print(total_loss)
    # Calculate the average loss for this epoch
    scheduler.step()
    train_loss = train_total_loss / len(train_loader)
    train_loss_history.append(train_loss)
        #print('toatl loss: ', total_loss)
        #print(len(train_loader))
        #print('average loss: ', average_loss)

    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {train_loss:.4f}')

    model.eval()
    with torch.no_grad():
        val_total_loss = 0.0
        for batch_X, batch_y in val_loader:
            predictions = model(batch_X)
            loss = criterion(predictions, batch_y)
            val_total_loss += loss.item()
        validation_loss = val_total_loss / len(val_loader)
        val_loss_history.append(validation_loss)

        # 检查是否早停
    if early_stopping(val_loss_history, patience = 30, delta = 0.001):
        print("\n Early stopping triggered!")
        best_model_state = model.state_dict()
        break
# 完成训练后，可以使用模型进行预测或进行评估。

# 使用训练好的模型进行预测

model.eval()
with torch.no_grad():
    val_total_loss = 0.0
    for batch_X, batch_y in val_loader:
        predictions = model(batch_X)
        loss = criterion(predictions, batch_y)
        val_total_loss += loss.item()
    validation_loss = val_total_loss / len(val_loader)    
    print(validation_loss)
    
    
```

```python
import torch
import torch.nn as nn

class MultiInputSingleOutputRNN(nn.Module):
    def __init__(self, input_size1, input_size2, hidden_size1, hidden_size2, output_size, time_step, num_layers1, num_layers2, dropout):
        super(MultiInputSingleOutputRNN, self).__init__()
        
        self.time_step = time_step
        
        self.hidden_size1 = hidden_size1
        self.hidden_size2 = hidden_size2
        
        self.num_layers1 = num_layers1
        self.num_layers2 = num_layers2
        

        self.rnn1 = nn.RNN(input_size1, hidden_size1, num_layers1, batch_first=True, dropout = dropout)
        self.rnn2 = nn.RNN(input_size2, hidden_size2, num_layers2, batch_first=True, dropout = dropout)
        self.fc1 = nn.Linear( (hidden_size1+hidden_size2) * self.time_step, output_size)        
        #self.fc2 = nn.Linear((hidden_size1+hidden_size2), output_size)

    def forward(self, input1, input2):
        batch_size = input1.size(0)
        sequence_length = input1.size(1)

        # Process input1 with rnn1
        hidden1 = self.init_hidden(self.num_layers1, batch_size, self.hidden_size1)
        out1, hidden1 = self.rnn1(input1, hidden1)
        #print(out1.shape)
        
        # Process input2 with rnn2
        hidden2 = self.init_hidden(self.num_layers2, batch_size, self.hidden_size2)
        out2, hidden2 = self.rnn2(input2, hidden2)
        #print(out2.shape)

        # Concatenate the outputs of rnn1, rnn2, and rnn3 along the feature dimension
        combined_output = torch.cat((out1, out2), dim=2)
        #print(combined_output.shape)
        
        out = combined_output[:, -self.time_step:, :]  
        #print(out.shape)
        
        #out = out.view(batch_size, -1)
        out = out.view(out.size(0), -1)
        #print(out.shape)
        
        out = self.fc1(out)
        
        #out = self.fc2(out)

        return out

    def init_hidden(self, num_layers, batch_size, hidden_size):
        return torch.zeros(num_layers, batch_size, hidden_size)
```
